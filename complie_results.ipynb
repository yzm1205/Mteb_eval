```python
import json
import numpy as np
mteb_score={}
for task in task_datasets:
    mteb_score[task] = []
    for dataset in save_dataset_list:
        dataset = dataset.split(".json")[0]
        if dataset in task_datasets[task]:
            with open(f"/data/yash/mteb/results/Bridge-AI_OpenELM-3B-Base-MTEB/no_model_name_available/no_revision_available/{dataset}.json") as f:
                data = json.load(f)
                print(data["scores"].keys())
                try:
                    mteb_score[task].append(data["scores"]["test"][0]["main_score"])
                except KeyError:
                    mteb_score[task].append(data["scores"]["test.full"][0]["score"])
    mteb_score[task] = np.mean(mteb_score[task])
```
